## VAE

```python
df = pd.read_csv('latent-vae.csv', index_col = 0)
```

```python
data = df.iloc[:, :11]
pc = PC(CItest = 'gaussian', alpha = 0.05, verbose=False).create_graph_from_data(data)
```

```python
plt.figure(figsize = (10, 10))
nx.draw(pc, with_labels=True, font_weight='bold')
```
